```python
#Load Data
import pandas as pd

train=pd.read_csv('E://Python Code/For Thesis/Data Sets/Adults/adult.data')
train.head()
```

```python
#Add Columns name of those data
train.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education_num',
                    'marital_status', 'occupation', 'relationship',
                    'race', 'sex', 'capital_gain', 'capital_loss',
                    'hours_per_week', 'native_country', 'income']
```

```python
#See the conditions after adding label 
train.head()
```

```python
import numpy as np

train= train.replace(' ?', np.nan).dropna()
```

```python
train.info()
```

```python
# For preprocessing the data
from sklearn.preprocessing import Imputer
from sklearn import preprocessing
# split X and y into training and testing sets
from sklearn.model_selection import train_test_split
```

```python
train_rev = train
```

```python
le = preprocessing.LabelEncoder()
workclass_cat = le.fit_transform(train.workclass)
education_cat = le.fit_transform(train.education)
marital_cat   = le.fit_transform(train.marital_status)
occupation_cat = le.fit_transform(train.occupation)
relationship_cat = le.fit_transform(train.relationship)
race_cat = le.fit_transform(train.race)
sex_cat = le.fit_transform(train.sex)
native_country_cat = le.fit_transform(train.native_country)
income_cat=le.fit_transform(train.income)
```

```python
#initialize the encoded categorical columns
train_rev['workclass_cat'] = workclass_cat
train_rev['education_cat'] = education_cat
train_rev['marital_cat'] = marital_cat
train_rev['occupation_cat'] = occupation_cat
train_rev['relationship_cat'] = relationship_cat
train_rev['race_cat'] = race_cat
train_rev['sex_cat'] = sex_cat
train_rev['native_country_cat'] = native_country_cat
train_rev['income_cat'] = income_cat
```

```python
#drop the old categorical columns from dataframe
dummy_fields = ['workclass', 'education', 'marital_status', 
                  'occupation', 'relationship', 'race',
                  'sex', 'native_country','income']
train_rev = train_rev.drop(dummy_fields, axis = 1)
```

```python
train_rev = train_rev.reindex_axis(['age', 'workclass_cat', 'fnlwgt', 'education_cat',
                                    'education_num', 'marital_cat', 'occupation_cat',
                                    'relationship_cat', 'race_cat', 'sex_cat', 'capital_gain',
                                    'capital_loss', 'hours_per_week', 'native_country_cat', 
                                    'income_cat'], axis= 1)

train_rev.head(20)
```
